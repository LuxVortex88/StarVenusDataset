```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Simulate star data from Venus orbit
np.random.seed(42)
n = 100
spectral_classes = ['O', 'B', 'A', 'F', 'G', 'K', 'M']
temps = {'O': 30000, 'B': 20000, 'A': 8500, 'F': 6500, 'G': 5500, 'K': 4500, 'M': 3000}

data = {
    'star_id': [f'VNS-{i:04d}' for i in range(n)],
    'spectral_type': np.random.choice(spectral_classes, n),
    'apparent_magnitude': np.round(np.random.uniform(-1, 10, n), 2),
    'distance_ly': np.round(np.random.uniform(4, 1000, n), 1)
}

df = pd.DataFrame(data)
df['temperature_K'] = df['spectral_type'].map(temps)

# Display first few rows
print(df.head())

# Plot: Apparent Magnitude vs Temperature
plt.figure(figsize=(8, 5))
plt.scatter(df['temperature_K'], df['apparent_magnitude'], c='orange', alpha=0.7)
plt.gca().invert_yaxis()  # Bright stars have lower magnitude
plt.xlabel('Temperature (K)')
plt.ylabel('Apparent Magnitude')
plt.title('Simulated Star Observations from Venus Orbit')
plt.grid(True)
plt.show()
```
